```python
import json
from urllib import request
import codecs
import numpy as np
import pandas as pd
```

```python
#Facebook API token
access_token = input('Enter access token: ')

#urls to check. no commas!
urls = input('Enter urls (no commas): ')

#split the input string on spaces
input_list_all = urls.split() 

#split one list into multiple lists by 50 urls in it
input_list_50 = [input_list_all[x:x+50] for x in range(0, len(input_list_all), 50)]
```

```python
#creating URLs for Facebook graph API

fb_urls = []
for i in input_list_50:
    urls=",".join(i)
    url_input = "https://graph.facebook.com/v2.8/?ids="+urls+'&access_token='+access_token
    fb_urls.append(url_input)
```

```python
#create one single file with the data from Facebook

reader = codecs.getreader("utf-8")
for i in fb_urls:
    response = request.urlopen(i)
    obj.update(json.load(reader(response)))
```

```python
#

#read_column = input('Enter access token: ')

myList = []
for key, value in obj.items():
    try:
        myList.append([key,
                       value['og_object']['title'],
                       value['og_object']['description'],
                       value['share']['comment_count'],
                       value['share']['share_count']
                      ])
    except KeyError:
        myList.append([key,"NaN"])
mf = pd.DataFrame(myList).set_index(0)
mf.index.names = ['Site']
mf.columns = ['title','description','comment count','share count']
mf
```

```python
mf.to_excel('facebook_score.xlsx', encoding='utf-8')
```
